### 六、合并与拼接
1. 横向合并：

   df1 = pd.DataFrame ({"id":[1,2], "name":["X","Y"]})

   df2 = pd.DataFrame ({"id":[1,2], "age":[20,30]})

   按 id 合并

```python
df1 = pd.DataFrame ({"id":[1,2], "name":["X","Y"]})
df2 = pd.DataFrame ({"id":[1,2], "age":[20,30]})
df3=pd.merge(df1,df2,on='id')
df3
```

2. 纵向拼接：

   df_a = pd.DataFrame ({"val":[1,2]})

   df_b = pd.DataFrame ({"val":[3,4]})

   拼接后重置索引。

```python
df3=pd.concat([df_a,df_b],ignore_index=True)
df3
```

### 七、时间处理
1. 创建时间列：["2025-01-01","2025-01-02","2025-01-03"]，转为 datetime 格式

```python
df1=pd.DataFrame({'dt':["2025-01-01","2025-01-02","2025-01-03"]})
df1['dt']=pd.to_datetime(df1['dt'])
```

2. 提取时间列的月份、星期。

```python
df1['年']=df1['dt'].dt.year
df1['月']=df1['dt'].dt.month
df1['日']=df1['dt'].dt.day
df1['星期']=df1['dt'].dt.day_name()
df1['周几']=df1['dt'].dt.weekday
df1
```


### 八、高级操作

1. 对 score 列做标准化（(值 - 均值)/ 标准差）。

```python
df1=pd.DataFrame({
    'name':['Tom','Jerry','Mike'],
    'score':[85,92,78],
    'gender':['男','女','男']
})
mean=df1['score'].mean()
std=df1['score'].std()
df1['标准化成绩']=(df1['score']-mean)/std
df1
```

2. 用透视表：行是 gender，值是 score，聚合函数为中位数。

```python
pd.pivot_table(
    data=df1,
    index='gender',
    values='score',
    aggfunc=['median','max','mean']
)
```
