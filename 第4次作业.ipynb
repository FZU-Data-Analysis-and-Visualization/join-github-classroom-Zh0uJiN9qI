```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.sans-serif']='SimHei'
plt.Figure(figsize=(10,5))
month1=['1月','2月','3月','4月','5月','6月','7月','8月','9月','10月','11月','12月']
sales1=[100,120,80,150,120,110,90,78,95,100,110,120]
month2=['1月','2月','3月','4月','5月','6月','7月','8月','9月','10月','11月','12月']
sales2=[50,70,80,30,80,100,60,55,67,87,29,40]
plt.plot(month1,sales1,label='产品A' ,color='orange',linewidth=2,linestyle='--',marker='*',ms=10)
plt.plot(month2,sales2,label='产品B' ,color='red',linewidth=2,linestyle='-',marker='o',ms=10)
plt.legend(loc='upper left')
plt.title('2026年销售额',color='blue',fontsize=20)
plt.xlabel('月份',fontsize=10)
plt.ylabel('销售额（万元）',fontsize=10)
plt.grid(True,alpha=0.2,color='blue',linestyle='--')
plt.xticks(rotation=0,fontsize=12)
plt.yticks(rotation=0,fontsize=12)
plt.ylim(0,160)
for x,y in zip(month1,sales1):
    plt.text(x,y+1,str(y),ha='center',va='bottom',fontsize=10)
for x,y in zip(month2,sales2):
    plt.text(x,y+1,str(y),ha='center',va='bottom',fontsize=10)
plt.tight_layout()
plt.show()
```
