练习题 1：数组创建与属性查看
题目： 使用 arange() 函数创建一个从 1 到 12 的一维等差数组，然后使用 reshape() 方法将其重塑为 3行4列的二维数组 。请分别打印出该数组的维度（ndim）、形状（shape）和元素总个数（size）。

练习题 2：数据类型转换
题目： 创建一个包含浮点数 [1.5, 2.6, 3.7, 4.8] 的 NumPy 数组。使用 astype() 方法将该数组的数据类型转换为 32位整数 (np.int32)，并打印转换前后的数组及其数据类型名称 (dtype.name) 。

```python
import numpy as np

float_arr = np.array([1.5, 2.6, 3.7, 4.8])

print(float_arr)
print(float_arr.dtype.name)

int_arr = float_arr.astype(np.int32)

print(int_arr)
print(int_arr.dtype.name)
```

练习题 4：花式索引与布尔索引
题目： 给定一维数组 arr = np.array([10, 15, 20, 25, 30, 35])。
1. 使用花式索引提取第 1、3、5 个元素（即索引为 0, 2, 4 的元素）。
2. 使用布尔索引提取数组中所有大于 20 的元素 。

```python
import numpy as np

arr = np.array([10, 15, 20, 25, 30, 35])

fancy_result = arr[[0, 2, 4]]

bool_result = arr[arr > 20]
(fancy_result,bool_result)
```

练习题 5：数组运算与广播机制
题目： 创建一个形状为 (3, 4) 的全 0 数组（使用 zeros() 函数）和一个一维数组 arr_b = np.array([1, 2, 3, 4]) 。利用 NumPy 的广播机制，将这两个数组相加并打印出结果 。

练习题 6：通用函数的应用
题目： 给定数组 x = np.array([1, 4, 9, 16]) 和 y = np.array([2, 1, 4, 5])。
1. 使用一元通用函数 np.sqrt() 计算数组 x 中每个元素的平方根 。
2. 将得到的平方根数组与数组 y 结合，使用二元通用函数 np.maximum() 找出对应位置上的最大值 。

```python
import numpy as np

x = np.array([1, 4, 9, 16])
y = np.array([2, 1, 4, 5])

sqrt_x = np.sqrt(x)

max_result = np.maximum(sqrt_x, y)
(sqrt_x, max_result)
```

练习题 7：数组的转置
题目： 创建一个形状为 (2, 3, 4) 的三维数组 arr = np.arange(24).reshape((2, 3, 4)) 。使用 transpose() 方法将该数组的轴顺序转换为 (1, 0, 2)，并打印转换后新数组的形状（shape） 。

```python
import numpy as np

arr = np.arange(24).reshape((2, 3, 4))

new_arr = arr.transpose(1, 0, 2)

(arr.shape,new_arr.shape)
```

练习题 8：条件逻辑运算
题目： 给出两个一维数组 arr_x = np.array([1, 5, 7])，arr_y = np.array([2, 6, 8]) 以及一个布尔条件数组 arr_con = np.array([True, False, True]) 。请使用 np.where() 函数，当条件为 True 时取 arr_x 中的对应元素，当条件为 False 时取 arr_y 中的对应元素 。

```python
import numpy as np

arr_x = np.array([1, 5, 7])
arr_y = np.array([2, 6, 8])
arr_con = np.array([True, False, True])

result = np.where(arr_con, arr_x, arr_y)

result
```

练习题 9：统计运算与排序
题目： 创建二维数组 arr = np.array([[6, 2, 7], [3, 6, 2], [4, 3, 2]]) 。
1. 使用统计方法计算该数组所有元素的算术平均值（mean()）。
2. 使用 sort() 方法对该数组沿着 axis=0 的方向（按列顺序从小到大）进行排序，并打印排序后的结果 。

```python
import numpy as np

arr = np.array([[6, 2, 7], [3, 6, 2], [4, 3, 2]])

mean_val = arr.mean()

arr_sorted = arr.copy()
arr_sorted.sort(axis=0)
(arr_sorted,mean_val)
```

练习题 10：随机数生成与矩阵点积
题目： 1. 设置随机数种子为 10（使用 np.random.seed() 保证每次生成的随机数一致）。
1. 使用随机数模块生成两个形状为 (2, 2) 且服从 [0, 1) 均匀分布的随机矩阵 A 和 B（使用 np.random.rand()）。
2. 使用线性代数模块的 dot() 方法计算矩阵 A 和矩阵 B 的点积 。

```python
import numpy as np

np.random.seed(10)

A = np.random.rand(2, 2)
B = np.random.rand(2, 2)

dot_result = np.dot(A, B)
(A,B,dot_result)
```
